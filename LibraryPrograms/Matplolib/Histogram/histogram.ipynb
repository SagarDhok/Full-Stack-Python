```python
#Implemetation of Histogram
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
marksvalues = np.random.randint(10,51,35)
print("Marks values ",marksvalues)
marksinterval =  np.array([10,20,30,40,50])
# plt.hist(marksvalues,marksinterval)
plt.hist(marksvalues,marksinterval,histtype="stepfilled")
plt.show()
```

```python
import numpy as np,pandas as pd, matplotlib.pyplot as plt
marksvalues = np.random.randint(10,51,35)
print("Marks values ",marksvalues)
marksinterval =  np.array([10,20,30,40,50])
plt.hist(marksvalues,marksinterval,histtype="bar",rwidth=0.7,edgecolor = "red")
plt.show()
```
