```python
import pandas as pd,matplotlib.pyplot as plt
year=["2021","2022","2023","2024"]
cse=[80,95,65,50]
it=[30,25,20,15]
ece=[12,15,6,18]
eee=[15,20,10,25]

ece_st = [cse[i]+it[i] for i in range(len(year))]
eee_st = [cse[i]+it[i]+ece[i] for i in range(len(year))]

plt.bar(year,cse,color="green")
plt.bar(year,it,bottom=cse,color="red")
plt.bar(year,ece,bottom=ece_st,color="orange")
plt.bar(year,ece,bottom=eee_st,color="purple")
plt.show()
```

```python
import pandas as pd,matplotlib.pyplot as plt
year=["2021","2022","2023","2024"]
cse=[80,95,65,50]
it=[30,25,20,15]
ece=[12,15,6,18]
eee=[15,20,10,25]

ece_st = [cse[i]+it[i] for i in range(len(year))]
eee_st = [cse[i]+it[i]+ece[i] for i in range(len(year))]

plt.bar(year,cse,color="green",label="CSE")
plt.bar(year,it,bottom=cse,color="red",label="IT")
plt.bar(year,ece,bottom=ece_st,color="orange",label="ECE")
plt.bar(year,ece,bottom=eee_st,color="purple",label="EEE")
plt.legend()
plt.show()
```
