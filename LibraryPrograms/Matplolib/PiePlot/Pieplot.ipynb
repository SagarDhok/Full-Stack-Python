```python
import matplotlib.pyplot as plt
games=[10,5,5,10,10]
labs = ["football","hockey","cricket","basketball","badminton"]
plt.pie(games,labels=labs,explode=[0.04,0,0,0,0])   #multiple pan explode karu shakto
plt.show()
```

```python
import matplotlib.pyplot as plt
games=[10,5,5,10,10]
labs = ["football","hockey","cricket","basketball","badminton"]
plt.pie(games,labels=labs,explode=[0.04,0,0,0,0],shadow=True)   #multiple pan explode karu shakto
plt.show()
```

```python
import matplotlib.pyplot as plt
games=[10,5,5,10,10]
labs = ["football","hockey","cricket","basketball","badminton"]
plt.pie(games,labels=labs,explode=[0.04,0,0,0,0],colors=["black", "hotpink", "b", "red","yellow"])   #multiple pan explode karu shakto
plt.legend(title="Game")
plt.show()
```

```python
import matplotlib.pyplot as plt
games=[10,5,5,10,10]
labs = ["football","hockey","cricket","basketball","badminton"]
plt.pie(games,labels=labs,explode=[0.04,0,0,0,0],colors=["green", "hotpink", "b", "red","yellow"],
        autopct="%0.0f",wedgeprops={"linewidth":4,"edgecolor":"skyblue"})   
plt.title("Games View",color="orange",fontweight="bold",fontsize=20)
plt.show()
```

```python
#============================================x==================================================
#Question: Priya lists down her monthly expenditure as follows:
#
#			Expenditure	Amount
#			Rent		4000
#			Food		5400
#			Clothing	2800
#			Savings	400
#		Draw a pie chart for her monthly expenses.
from matplotlib import pyplot as plt
plt.title("Expenditure",color="green",fontweight="bold",fontsize=20)
exprs = [4000,5400,2800,1400]
exnames = ["Rent","Food","Clothing","Savings"]
cols=['rosybrown', 'moccasin', 'lightyellow', 'darkseagreen']
plt.pie(exprs,labels=exnames,explode=[0.1,0,0,0],colors=cols,autopct="%0.0f",
        wedgeprops={"linewidth":2,"edgecolor":"red"}
        )
plt.show()
```
