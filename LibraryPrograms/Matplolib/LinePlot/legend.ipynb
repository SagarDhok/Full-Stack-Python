```python
import matplotlib.pyplot as plt,numpy as np
x = np.arange(1,6)
y = x**2
z = x**3
plt.plot(x,x,marker ="o",c="r")
plt.plot(x,y,marker = "X",c="g")
plt.plot(x,z,marker="*",c="black")
lab = ["NatNumSum","NatSumSquare","NatNumCubes"]
# plt.legend(labels = lab)
#*or loc means position of fancy box or legend 
plt.legend(labels = lab,loc= "best")
plt.show()

#! bus itparyant liha an position change karu shakto  
```
