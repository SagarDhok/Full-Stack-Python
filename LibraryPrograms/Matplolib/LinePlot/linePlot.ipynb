```python
#3. Setting Labels for X-axis and Y-Axis and Title.
#Setting the Markers Properties
#Draw a plot with the following Value
from matplotlib import pyplot as plt
x = [1,2,3,4]
y = [1,2,3,4]
plt.plot(x,y,ls="solid",lw=6,marker="o",mec= "green",mfc="yellow",ms=20,c="red")
plt.xlabel("X-axis") 
plt.ylabel("Y-axis")
plt.title("Line Graph")
plt.show()
```

```python
#Set the fonf Properties
from matplotlib import pyplot as plt
import numpy as np
x = np.array([1,2,3,4])
y = x**2
plt.plot(x,y,ls="solid",lw=6,c="r",marker="o",mec="green",mfc="yellow",ms=10)
font1 = {"family":"serif","color":"blue","size":16}
plt.xlabel("X-axis",fontdict=font1)
plt.ylabel("Y-axis",fontdict=font1)
plt.title("Line Graph")
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
x = np.array([1,2,3,4])
y = x**2
plt.plot(x,y,ls="solid",lw= 6,c="r",marker="o",mec="green",mfc="yellow",ms=15)
#Set the fonf Properties
font1  = {"family":"serif","color":"blue","size":16}
font2 = {"family":"arial","color":"blue","size":16}
font3 = {"family":"times new roman","color":"red","size": 16}
plt.xlabel("X-axis",fontdict=font1)
plt.ylabel("Y-axis",fontdict=font2)
plt.title("Line Graph",fontdict=font3)
plt.show()
```

```python
#How to plot Multiple Plots (Lines) in same axes
import matplotlib.pyplot as plt,numpy as np
x = np.arange(1,6)
y = x**2
plt.plot(x,x,marker="o",c="r")
plt.plot(x,y,marker="X",c="g")
plt.show()
```

```python
#How to plot Multiple Plots (Lines) in same axes
import matplotlib.pyplot as plt,numpy as np
x = np.arange(1,6)
y = x**2
z = x**3
plt.plot(x,x,marker = "o",c="r")
plt.plot(x,y,marker ="X",c="g" )
plt.plot(x,z,marker = "*",c="black")
plt.show()
```
