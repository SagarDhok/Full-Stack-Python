```python
#Drawing Sub Plotting
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.plot([1,2,3,4],[4,3,2,1])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("XY plot-1")
print("--------------------")
plt.subplot(1,2,2)
plt.plot([1,2,3,4],[1,2,3,4])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("XY-plot-2")
print("--------------------")
plt.suptitle("Main plot contains subplots")
plt.show()
```

```python
# Implement the above Problem with Multiple Sub Plots in same axes
import pandas as pd,matplotlib.pyplot as plt
df= pd.read_csv("E:\\FULL STACK PYTHON\\LibraryPrograms\\tempweek.csv")
print(df)
# Draw a Sub Plot for days VS week1
plt.subplot(3,1,1)
plt.plot(df["Days"],df["Week1"],lw=6,c="red")
plt.title("First Week",fontdict={"family":"arial","color":"blue","size":16})

# # #Draw a Sub Plot for days VS week2
plt.subplot(3,1,2)
plt.plot(df["Days"],df["Week2"],lw=6,c="green")
plt.title("Second Week",fontdict={"family":"arial","color":"blue","size":16})

# # #Draw a Sub Plot for days VS week3
plt.subplot(3,1,3)
plt.plot(df["Days"],df["Week3"],lw=6,c="black")
plt.title("Third Week",fontdict={"family":"arial","color":"blue","size":16})

# # #Set the super title
plt.suptitle("Temp in 3 WEEK DAYS",fontdict={"family":"arial","color":"red"})
plt.show()


```

```python
import numpy as np
import pandas as pd
rm=pd.DataFrame({"Month":["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
                        "Sales_Prod_1": np.random.randint(50,200,12),
                        "Sales_Prod_2": np.random.randint(60,180,12)})
print("Reatil Company Data")
print(rm)
import matplotlib.pyplot as plt
plt.plot(rm["Month"],rm["Sales_Prod_1"],lw=3,c="red",marker="o")
plt.plot(rm["Month"],rm["Sales_Prod_2"],lw = 3, c="green",marker="x")
plt.xlabel("Months",fontdict={"family":"arial","color":"blue","size":16})
plt.ylabel("Sales Trend",fontdict={"family":"arial","color":"blue","size":16})
plt.title("Sales in Months",fontdict={"family":"arial","color":"blue","size":20})
plt.legend(labels=["Sales_Prod_1","Sales_Prod_2"])
plt.grid(axis="both",ls="dashed",c="green",lw=2)
plt.show()
```
