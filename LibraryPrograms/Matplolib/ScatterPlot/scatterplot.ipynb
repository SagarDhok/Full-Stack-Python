```python
import numpy as np,matplotlib.pyplot as plt,pandas as pd
d = {"Carage":[5,7,8,7,2,17,2,9,4,11,12,9,6],
     "CarSpeed":[99,86,87,88,111,86,103,87,94,78,77,85,86]
     }
df= pd.DataFrame(d)
print(df)
plt.scatter(df["Carage"],df["CarSpeed"],color="red",s=50,marker="D")
plt.show()
```

```python
import matplotlib.pyplot as plt,numpy as np
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange",
                   "purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x,y,c=colors)
plt.show()
```

```python
#dots cha and cmap colorbar cha color same hote
import matplotlib.pyplot as plt,numpy as np
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.scatter(x,y,c=colors,cmap="viridis")
plt.colorbar() #color bar enable karte
plt.show()
```

```python
#dots cha and cmap colorbar cha color same hote
import matplotlib.pyplot as plt,numpy as np
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y,c=colors,cmap="rainbow",s=sizes)
plt.colorbar() 
plt.show()
```

```python
#----------------------
#Problems
#-----------------------
#Let us understand how to construct a scatter plot with the help of the below example.
#Draw a scatter plot for the given data that shows the number of games played and scores obtained in each instance.
#No. of games	3	5	2	6	7	1	2	7	1	7
#Scores		80	90	75	80	90	50	65	85	40	100
import matplotlib.pyplot as plt,numpy as np,pandas as pd

d={"No_of_games":[3,5,2,6,7,1,2,7,1,7],
    "Scores":[80,90,75,80,90,50,65,85,40,100]}
df = pd.DataFrame(d)
print(df)
```

```python
plt.scatter(df["No_of_games"],df["Scores"])
plt.show()
```

```python
#Problems
#-----------------------
#Every year, a meteorologist studying tropical storms predicts how many hurricanes will occur. 
#The table below shows how many hurricanes were predicted and how many hurricanes actually occurred for the years 
# 2000 to 2005

 
#Year:		2000	2001	2002       2003       2004    2005
#Predicted	11		13		12		    17		   15	   17-------------------x axis
#Actual		14		15		12	        16		   14	   27-------------------y-axis
import matplotlib.pyplot as plt,numpy as np
d={"Predicted":[11,13,12,17,15,17],
    "Actual":[14,15,12,16,14,27]}
df = pd.DataFrame(d)
print(df)
plt.scatter(df["Predicted"],df["Actual"])
plt.show()
```

```python
#Problem
#------------
#Rossum is curious how different amounts of sunlight will impact the growth of his tomatoes. 
#he exposed some of his tomatoes to varying amounts of sunlight (in hours per day) and recorded the change 
#in their circumferences (in centimeters) after a week. His data is summarized in the table below.
#		Sunlight (hours per day)	: 0     2    4       6      8
  
#     Change in circumference (cm)	:1.00  2.50 5.25  10.75   11.25
import matplotlib.pyplot as plt,numpy as np,pandas as pd
Sunlight = pd.Series([0,2,4,6,8])
circumference = pd.Series([1.00,2.50,5.25,10.75,11.25]) 
print(Sunlight)
print(circumference)

plt.scatter(Sunlight,circumference)
plt.show()
```
