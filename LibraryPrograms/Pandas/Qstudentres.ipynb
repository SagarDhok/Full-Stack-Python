```python
# 6.find the total number of students who passed in disctintion (before solving this quition calculate total marks percentage and grade)
df["totalmarks"]=df["english"]+df["hindi"]+df["maths"]+df["science"]+df["social"]+df["telugu"]
df["percentage"]=round(df["totalmarks"]/600*100,2)
df["grade"] = None
print(df)
```

```python
df.loc[df["percentage"]>74,["grade"]] = "Distinction"
df.loc[((df["percentage"]<74) & (df["percentage"]>=65)) ,["grade"]] = "First"
df.loc[((df["percentage"]>=60) & (df["percentage"]<65)) ,["grade"]] = "Second"
print(df)
```
