```python
# 14. How many people have a credit card that expires in 2020?

#Step:1 Retrive all the Column Names
df.columns
```

```python
#Step:2 Retrive the Column that we want to work 
df['CC Exp Date']
```

```python
#Step:3 Creating a function to find the exp year of Credit Card
def findexp():
    cnt=0
    for date in df['CC Exp Date']:
        if date.split('-')[1]=='20':
            cnt=cnt + 1
    return cnt
```

```python
#Step:4 Finding the Number of people that having their Credit card expiry in 2020
print('Number of people having their Credit card expiry in 2020:',findexp())
```
